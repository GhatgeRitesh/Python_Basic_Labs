```python
# Using the Variable to Store the String of customer review
review="Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."

#Creating the class to define the methods related to TextAnalyzer Functioning
class TextAnalyzer (object):
    fmt=""
    def __init__(self,text):
        # removing punctuations 
        formatedtext=text.replace('.','').replace('!','').replace('?','').replace(',','')
        # converting to lowercase 
        formatedtext.lower()
        #setting formatedtext to class variable 
        self.fmt=formatedtext

    def freqAll(self):
        #spliting the text into list 
        list=self.fmt.split()
        #count the frequency
        map={} #Declarign the Dictionary to set the word and count as key value pair 
        for word in set(list): #using the set to remove duplicates from the list
            map[word]=list.count(word)
            
        return map

    def freqOf(self,word):
        mapDic=self.freqAll()
        if word in mapDic:
            return mapDic[word]

        return 0
        
# Creating the instance of class textanalyzer 
TA=TextAnalyzer(review)
print(TA.freqAll())
print("frequnecy of \"diam\":",TA.freqOf("diam"))
        
```
